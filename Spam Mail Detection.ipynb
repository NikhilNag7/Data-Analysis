```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
raw_mail_data = pd.read_csv("/content/mail_data.csv")
```

```python
raw_mail_data
```

```python
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
mail_data.head()
```

```python
mail_data.shape
```

```python
mail_data.loc[mail_data['Category'] == 'spam', 'Category',]= 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',]= 1
```

```python
mail_data.head()
```

```python
X = mail_data['Message']
Y = mail_data['Category']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state =3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df = 1,stop_words = 'english',lowercase='True')
```

```python
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

```python
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features,Y_train)
```

```python
prediction_on_training_data = model.predict(X_train_features)
prediction_on_accuracy_score = accuracy_score(Y_train,prediction_on_training_data)
```

```python
print('Accuracy on Training data:'" ",prediction_on_accuracy_score)
```

```python
prediction_on_test_data = model.predict(X_test_features)
prediction_on_accuracy_score = accuracy_score(Y_test,prediction_on_test_data)
```

```python
print('Accuracy on Testing data:'" ",prediction_on_accuracy_score)
```
